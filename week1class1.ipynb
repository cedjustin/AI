```python
#problem 1
print("problem 1")
import matplotlib.pyplot as plt

# a constant of the first grain they gave him
GRAIN = 1

# a list to contain the number of grains he got every day
grains_per_day = []

# a loop to calculate the number of grains in 100 days
for day in range(100):
  if day == 0:
    grains_per_day.append(GRAIN)
  else:
    grains_per_day.append((GRAIN*2)**day)

print("On the 100th day he had {:,} grains of rices".format(sum(grains_per_day)))

# the creation of the line graph
plt.title ( "​ ​Grains of rices ​" ) 
plt.xlabel ( "​ ​number of days ​" ) 
plt.ylabel ( "​ number of grains​ ​") 
plt.plot(grains_per_day, linestyle='dashed', linewidth=2, color='green')
plt.show()
# end of problem 1
```
